## `With` environment for opening files

Last time we discussed working with files, but I didn't have time to tell you about an important construction `with`, which is often used to automatically close open files. Let's consider the application of this construction by example.

First, let's create a file that we will open. Let it be `test_123.py`. If you have a file with this name in your laptop folder that is very valuable to you, then replace the value of the `filename` variable with something else.

```python
# set the filename
filename = 'test_123.py'
```

```python
f = open(filename, 'w')              # open a file for recording
f.write("print('Hello, world!!!')")  # write a Python command to the file
f.close()                            # close the file "by hand"
```

The standard problem with files is that any open file must be closed. In the previous lecture I showed you this syntax:

```python
# open the file and read it
open(filename).read()
```

This is a very short, but not very good syntax, because closing of the file is left to the so-called garbage collector system, and it is not known exactly when the file will be closed.

*In the standard Python implementation, which is called CPython, garbage collector is designed in such a way that the file is closed immediately after executing this line, but other implementations may behave differently and in some situations problems may arise with such code.*

It's better to do this:

```python
# open file using "with" instruction
with open(filename) as f:
    print(f.read())
```

When entering the `with` construct, a line equivalent to `f = open(filename)` is executed. Then the indented lines are executed, and when the indentation ends, the file will be closed automatically. So these two lines are equivalent to:

```python
# similar file operations without "with"
f = open(filename)
print(f.read())
f.close()
```

Here is another examples:

```python
# open file using "with" instruction and read it twice
with open(filename) as f:
    print(f.read())       # read information from the file
    f.seek(0)             # go back to the beginning of the file
    print(f.read())       # read information from the file again
```

Here we use the construction `f.seek(0)` to "rewind" the file to the beginning — in this case, repeated `f.read()` will again output its contents.

Now let's try to do something with the file after the indentation.

```python
# open file using "with" instruction and read it twice
with open(filename) as f:
    print(f.read())       # read information from the file
    f.seek(0)             # go back to the beginning of the file
    print(f.read())       # read information from the file again
    
# attempt to read the file after closing it
print(f.read())
```

As you can see, immediately after the end of the block (indented, as usual), the file is closed.

The `with` syntax has several advantages over the traditional approach. Firstly, you certainly won't forget to close the file, because you can forget to write `f.close()`, but you can't forget to remove the indentation. Secondly, even if you don't forget `f.close()`, you may not reach it because there was some mistake on the way.

### A little bit about exceptions

In the code below, after the file has been opened, division by 0 occurs. Construction

<pre>
try:
    something
except Name_of_some_error:
    do_something_else
</pre>

allows, if an error of the `Name_of_some_error` type occurs, not to end the program with the words "Everything is gone! Error!", but immediately transfer control to the block `do_something_else`, which will do something. Interestingly, what is in the block
`do_something_else` in the example below, the file was still open, which is bad. This can be compared to the situation: you put the teapot on the cooker, but then you got an urgent call and you ran away, and the fire was left out.

```python
try:
    f = open(filename)                  # open the file
    print(f.read())                     # print data from file
    print(10/0)                         # try to divide on 0 (makes an error)
    print('This is never been printed') # commands after error (never happen)
    f.close()                           # close the file
except ZeroDivisionError:
    print("Ups, I did it again!")       # print a message inside a exception block
    f.seek(0)                           # move to the start of the file
    print(f.read())                     # print data from file again
```

And here the situation is like this: although we ran away on an urgent call, but the smart teapot immediately turned itself off. As you can see, when trying to read from a file in the `except` block, we get an error, and this is good, it means that the file has closed, despite the error.

```python
try:
    with open(filename) as f:               # open the file safely
        print(f.read())                     # print data from file
        print(10/0)                         # try to divide on 0 (makes an error)
        print('This is never been printed') # commands after error (never happen)
except ZeroDivisionError:
    print("Ups, I did it again!")           # print a message
    print(f.read())                         # try to print file data after closing the file
```

## Appending to a file

We often need to do one thing with a file - either read or write. Sometimes we need to modify a file. Most often, this is done as follows: the file is first read into memory, then modified in memory and written "from scratch" to the same place as before. If we are talking about not very large files, then this method works fine.

At the same time, sometimes we don't need to overwrite the file from scratch, but add some information to the end of the file. Most often, this has to be done to record logs in which some information about the operation of the program is stored (for example, this way the web server logs from which addresses it was accessed and which pages were requested). To add something to the end of the file, you need to open it with the modifier `'a'` (from the word `append`) like this:

```python
# write information to the end of the file
with open(filename, 'a') as f:
    print("\n" + "print('Some new string')", file = f)
```

Let's check that the old content is still in place:

```python
# check tupdated content of file
with open(filename) as f:
    print(f.read())
```

As you can see, it's okay.

### A little bit about HTML

What you see above is an HTML page. HTML (HyperText Markup Language) is such a markup language, which is a special case of the SGML standard. Another special case of SGML is XML, which we will meet again.

Let's write a simple HTML page. It is most convenient to do this in any editor. But I'll write it to a file via my notebook.

```python
# create a string containing an html page
my_html = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset = "UTF-8">
    <title>Title</title>
</head>
<body>
<h1>Hello</h1>
<p>I'm a paragraph.</p>
<hr>
<ol>
    <li>One</li>
    <li>Two</li>
</ol>

</body>
</html>
'''
```

```python
# write html string to the html file
with open('my.html', 'w') as f:
    f.write(my_html)
```

Open `my.html` with browser and you will see a simple web page. It can be seen that HTML is divided into special fragments, which are called tags. In the text above there are tags: `<html>`, `<head>`, `<title>`, etc. Each tag marks some piece of the web page. The `<title>` tag is the page title. The `<ol>` tag marks an ordered list. The `<li>` tag corresponds to the list item. The `<p>` tag is a paragraph. All the listed tags are *paired*: they mark some fragment of text (possibly containing other tags), placing it between the corresponding opening and the closing tag (for example, `<li>` is the opening tag, and `</li>` is the closing tag; everything in between is a list item). The exception here is the `<hr>` tag, which draws a horizontal line (it works without `</hr>`).

In fact, an HTML page is a set of nested tags. We can say that this is a tree with a root in the `<html>` tag. Each tag has descendants - those tags that are directly nested in it. For example, the `<body>` tag will have `<h1>`, `<p>`, `<hr>`, `<ol>` descendants. It turns out such a kind of family tree.

HTML interests us in order to extract information from such a tree. One of the most popular objects for storing information are tables, so let's insert a small table into our file: it is indicated by the `<table>` tag, each row of the table is marked with the `<tr>` tag inside `<table>`, and each cell is marked with the `<td>` tag inside `<tr>`.

```python
# create a string containing an html page with table
my_html = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset = "UTF-8">
    <title>Title</title>
    <style type='text/css;'>
        table {
        border-collapse: collapse;
    }
    table, th, td {
        border: 1px solid black;
    }
    </style>
</head>
<body>
<h1>Hello</h1>
<p>I'm a paragraph.</p>
<hr>
<ol>
    <li>One</li>
    <li>Two</li>
</ol>
<table>
    <tr>
        <td>
            Cell 1
        </td>
        <td>
            Cell 2
        </td>
    </tr>
    <tr>
        <td>
            Cell 3
        </td>
        <td>
            Cell 4
        </td>
    </tr>
</table>
</body>
</html>
'''

# write html string to the html file
with open('my.html', 'w') as f:
    f.write(my_html)
```

## BeautifulSoup

There are many packages for processing web pages. The problem with HTML is that most browsers behave "forgiving", and therefore there are a lot of poorly written (not according to the HTML standard) HTML pages on the web. However, processing even not quite correct HTML code is not so difficult if you have the right tools at hand.

We will use the *Beautiful Soup 4* package. It is included in the standard *Anaconda* package, but if you use another Python distribution, you may have to install it manually using `pip install beautifulsoup4`.

A package called `BeautifulSoup` is most likely not what you need. This is the third version (*Beautiful Soup 3*), and we will use the fourth. So we need a package `beautifulsoup4`. To make it really fun, when importing, you need to specify a different
package name — `bs4`, and import a function called `BeautifulSoup`. In general, it's easy to get confused at first, but these difficulties need to be overcome once, and then it will be easier.

```python
# import Beautiful Soup
from bs4 import BeautifulSoup
```

To use *BeautifulSoup*, you need to pass the text of the web page (as a single line) to the `BeautifulSoup` function. In order for him not to swear, I also manually specify the name of the parser (the program that just processes HTML) — for compatibility, I use `html.parser` (it is included in the Python package and does not require installation), but you can also try using `lxml` if you have it installed.

```python
# parse html with BeautifulSoup
page = BeautifulSoup(r.text, 'html.parser')
```

What now lies in the `page` variable? Let's take a look.

```python
page # print parsed document
```

We see that the `page` object is very similar to a string, but, in fact, it is not just a string. You can make requests to the `page`.

```python
# request to page: print html tag and its content
page.html
```

We see what is inside the `<html>` tag (this is almost the entire page, but the very first line was "cut off"). You can go deeper and look at the contents of `<head>`.

```python
# request to page: print head tag and its content
page.html.head
```

Now we only see what's inside the `<head>` tag. We can go even deeper and get what is inside the `<title>` tag, which in turn is inside the `<head>` tag (they say that `<title>` is a *descendant* of `<head>`):

```python
# request to page: print title tag and its content
page.html.head.title
```

However, it would be possible not to write in such detail — since there is only one `<title>` tag in the document, we could not specify that it is inside `<head>`, which is inside `<html>`.

```python
# request to page: print title tag and its content
page.head.title
```

```python
# request to page: print title tag and its content
page.title
```

One of the descendants of `<body>` is `<table>`. You can get it like this:

```python
# request to page: print table tag and its content
page.body.table
```

Let's say I need to get multiple items with the same tag, for example, all rows of `<tr>`. The following syntax is used for this:

```python
# get multiple items with the same tag <tr>
rows = page.body.table.findAll('tr')
rows
```

```python
# number of rows in 'rows'
len(rows)
```

We see that this is a list of two elements. So you can loop through it.

```python
# print all found rows and its indexes
for i, row in enumerate(rows, 1):
    print(i)
    print(row)
```

We have 2 lines and each of them is the same `BeautifulSoup` object as all the previous ones. So you can apply the `row.td` construction to them.

```python
# print first <td> in each found row
for i, row in enumerate(rows):
    print(i)
    print(row.td)
```

We see that if there are several `<td>` tags inside the `<row>` tag, then `row.td` will take the first of them. So we got the first column. But we are not interested in the `<td>` tag itself, but in the string that lies there. It can be printed like this:

```python
# print inner text of first cell of each found row
for i, row in enumerate(rows):
    print(i)
    print(row.td.string)
```

It can be seen that there are unnecessary spaces before the line. Delete them with the `strip` command.

```python
# print inner text of first cell of each found row without spaces 
for i, row in enumerate(rows):
    print(i)
    print(row.td.string.strip())
```

Let's load the table as a list of lists:

```python
table = []                                     # new empty list for storing the table data
for i, row in enumerate(rows):                 # run over each found row
    table.append([])                           # add an empty list as a new row to the table
    for cell in row.findAll('td'):             # run over each element found in row
        table[-1].append(cell.string.strip())  # add found element in the table
print(table)                                   # output the final table data
```

Here is the same thing, but shorter with the help of list comprehensions:

```python
# load data from HTML table to Python table (using list comprehension)
table = []                                     
for row in rows:          
    table.append([cell.string.strip() for cell in row.findAll('td')])
print(table)                           
```

Or even shorter (but more intricate):

```python
# load data from HTML table to Python table (using nested list comprehensions)
table = [[cell.string.strip() for cell in row.findAll('td')] 
         for row in rows]
print(table)
```

Note that instead of `some_beautiful_soup_object.findAll('sometag')` you can write shorter `some_beautiful_soup_object('sometag')`. So you can write even shorter:

```python
# load data from HTML table to Python table (shortest way)
table = [[cell.string.strip() for cell in row('td')] 
         for row in rows]
print(table)
```

Tags, in addition to the name, have other properties — for example, in the line `<html lang="en">` we see the `lang` property of the `<html>` tag, which has the value `"en"`. Another important example of a tag with properties is the `<a>` tag, which creates a link. It has a `href` property that stores the actual link.

Now imagine that we want to make a robot that will walk on web pages and go from one page to another by links. Then we are faced with the task of extracting all hyperlinks from the page. To do this, you need to find all the `<a>` tags on the page, and take the `href` parameter from all of them. To begin with, let's show how to get an object property, for example, `lang` from `html`. This is done as if our object is a dictionary, and we take its value by key.

```python
page.html['lang'] # get tag's property and print it
```

If we request a property that the tag does not have, then we will get a KeyError, as with the dictionary

```python
page.html['strange'] # try get non existent property
```

Just like the dictionary, there is a `get()` method that returns nothing if there is no such property. Or returns the default value defined by us.

```python
# use the get() method to check for the presence of a property
page.html.get('strange')
```

```python
# use the get() method to check for the presence of a property (with default value)
page.html.get('strange', 'no-such-tag')
```

Now we will extract all the links from some site.

```python
r = requests.get('http://www.sevsu.ru')     # get network request's result
page = BeautifulSoup(r.text, 'html.parser') # parse the result
```

Here are all the links on our page.

```python
page('a') # get all 'a' tags
```

As you can see, the `findAll()` method (or its abbreviated form of writing in the form of just parentheses) searches not only for the immediate "children" of some vertex (in genealogical terms), but also for all descendants.

Let's print the links ourselves:

```python
for link in page("a"):           # for each found link in parsed document
    if link.get("href") != None: # print URL if link has href attribute
        print(link["href"])
```

There are external hyperlinks that start with `http`, and local ones that lead to the same site and are relative (that is, before `1516/topology2.php` you need to write `http://www.sevsu.ru/` to get the full link to the relevant document).

Now it is clear how our robot should act: for each of the received links, it should load the corresponding page, find all the links on it, add them to the queue for research, etc. This is about how web crawlers of search engines work (although, of course, they are much more complicated).
